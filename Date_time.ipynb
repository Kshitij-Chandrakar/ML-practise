## Working with date 

```python
#Converting to date and time
date['date']=pd.to_datetime(date["date"])
```

```python
date.info()
```

### Extract Year 

```python
date["date_year"]=date["date"].dt.year
```

```python
date.head()
```

### Extract Month 


```python
date["date_month"]=date["date"].dt.month_name()
date.head()
```

### Extract days

```python
date["date_days"]=date["date"].dt.day
date.head()
```

### Week extraction

```python
date["dow"]=date["date"].dt.dayofweek
date.head()
```

### Day of week name 

```python
date["day_of_week"]=date["date"].dt.day_name()
date.head()
```

### Is weekend

```python
date["date_is_weekend"]=np.where (date["day_of_week"].isin(["Sunday","Sunday"]),1,0)
date.head()
```
