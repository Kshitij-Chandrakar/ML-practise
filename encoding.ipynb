### Train test split

```python
x=df.drop(columns=["has_covid"])
y=df["has_covid"]
```

```python
y.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_test.head()
```

### Column transformer 

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder,OneHotEncoder
from sklearn.impute import SimpleImputer #to fill the missing values 
```

```python
transformer = ColumnTransformer(transformers=[('tnf1',SimpleImputer(),['fever']),('tnf2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),('tnf3',OneHotEncoder(sparse_output=False,drop='first'),['gender','city'])],remainder='passthrough')
```

```python
x_train_new=transformer.fit_transform(x_train)
x_test_new=transformer.fit_transform(x_test)
```

```python
x_train_new
```

```python
x_train_new.shape
```

```python
x_test_new.shape
```

```python
x_train
```

```python
df
```

```python
feature_names = transformer.get_feature_names_out()
#After all your internal changes, what are the final column names?”
#That is exactly what get_feature_names_out() returns.
```

```python
feature_names
```

```python
X_train_df = pd.DataFrame(
    x_train_new,
    columns=feature_names,
    index=x_train.index
)

X_test_df = pd.DataFrame(
    x_test_new,
    columns=feature_names,
    index=x_test.index
)
```

 ### pd.DataFrame(X_train_new)

Converts array → table

Still has no column names

### columns=feature_names

Assigns correct transformed column names

Ensures each column has meaning

### index=X_train.index

Keeps row alignment

Ensures rows still match y_train
### If you skip index(last one):

Labels can silently misalign

This causes serious ML bugs
